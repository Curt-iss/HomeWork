### Part C

```python
# Define f(x, b) and the gradient of f(x, b)
def f(x, b):
    total = 0
    for i in range(len(x)):
        total += (x[i] - b[i])**2
    return total

def gradf(x, b):
    return 2 * (x - b)

# For Part C, I'll print the output of these functions when x is filled with ones

x1 = np.ones(100)
print("The output of f(x,b) when x is filled with one is: \n", f(x1, b))
print(" \n The output of gradf(x, b) when x is filled with ones is: \n", gradf(x1, b))
```

### Part F

```python
# Define Function approxDirectDeriv 
def approxDirectDeriv(x, b, d):
    # Find the magnitude for d
    mag = 0
    for di in range(len(d)):
        mag += d[di]**2
    mag = np.sqrt(mag)
    # return the given approximation formula
    return (f(x+d, b) - f(x, b))/mag
```

### Part G

```python
singleH = 0.0001
d = [-singleH] * 100
approx = approxDirectDeriv(a, b, d)
error = np.abs(approx - direcDeriv(a, d))
print("The approximate directional derivative with the giving information is:", approx)
print("The error of this approximation is: ", error)
```
